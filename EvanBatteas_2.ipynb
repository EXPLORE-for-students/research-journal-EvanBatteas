## Motivation:

We want to find Milky Way and Andromeda analogues in the COCO WDM local group simulation, and to do this, we have to select subhalos which have properties similar to the real life Milky Way and Andromeda. We want them to be of similar mass to real life, a similar distance away to real life, relativly isolated like real life, and moving towards each other like real life. The overall goal of the project is to use the kinematics of these analogues to make an estimate of the mass of the local group, via either the timing argument or virial theorem [1]. We can then comapre this mass estimate to the ones obtained when using a CDM simulation, seeing how WDM affects the estimate.
